```stata
use firm_green_entry.dta
```

```stata
gen depvar = (num_pat_ccmt_2>0)*1
```

```stata
gen nsrc = log(src+1)
```

```stata
gen npath = log(path+1)
```

```stata
sum npath,d
local L `r(p5)'
local H `r(p95)'
local STEP = (`H'-`L')/10
reghdfe depvar c.nsrc##c.npath, absorb (person_id y02) cluster(person_id y02) residuals nocons
margins, dydx(nsrc) at(npath=(`L'(`STEP')`H')) predict(xb)
marginsplot, recast(line) recastci(rarea) plotopts(lcolor(blue) lpattern(solid) lwidth(medium) fcolor(navy%30) ) name(margins_npath, replace) xlabel(#5) title("marginal effect of starting points") ytitle("{&part}log(#patents+1)/{&part}log(#starting points+1)") xtitle("log(#pathways+1)")  aspect(1) scheme(s1mono)
graph export "margins_nsrc.png", replace
graph export "margins_nsrc.pdf", replace
```

```stata
sum nsrc,d
local L `r(p5)'
local H `r(p95)'
local STEP = (`H'-`L')/10
reghdfe depvar c.nsrc##c.npath, absorb (person_id y02) cluster(person_id y02) residuals nocons
margins, dydx(npath) at(nsrc=(`L'(`STEP')`H')) predict(xb)
marginsplot, recast(line) recastci(rarea) plotopts(lcolor(blue) lpattern(solid) lwidth(medium) fcolor(navy%30)) name(margins_nsrc, replace)  xlabel(#5) title("marginal effect of pathways") ytitle("{&part}log(#patents+1)/{&part}log(#pathways+1)") xtitle("log(#starting points+1)") aspect(1) scheme(s1mono)
graph export "margins_npath.png", replace
graph export "margins_npath.pdf", replace
```
